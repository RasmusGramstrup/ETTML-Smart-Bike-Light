```python
import os
import csv
import pandas as pd

for file in os.listdir('./out/05_05'):
    count = 1
    with open('./out/05_05/' + file, 'r', encoding="utf-8") as inp:
        lines = list(csv.reader(inp))
        print(lines[1])
        

        while True:
            fileName = ' '.join(map(str,lines[1])) + '.' + format(count,'05d') + '.csv'
            filePath = "./out_converted"
            path = os.path.join(filePath, fileName)
            print(path)
            if os.path.exists(path):
                count += 1
                continue
            else:
                break
        inp.close()

        df = pd.read_csv('./out/05_05/' + file)
        df =  df.drop(df.index[0])
        df.to_csv(path, index=False)
        
        # with open('./out/05_05/' + file, 'r', encoding="utf-8") as inp, open(path, 'w', encoding="utf-8") as output:
        #     writer = csv.writer(output)
        #     for row in csv.reader(inp):
        #         if row == lines[1]:
        #             continue
        #         else:
        #             writer.writerow(row)
                   

    # inp.close()
    # output.close()
        
    
    
```
